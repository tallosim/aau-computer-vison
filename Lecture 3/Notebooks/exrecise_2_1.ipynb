```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from helpers import *
```

```python
image1 = cv2.imread('../Images/aau-city-1.jpg')
image2 = cv2.imread('../Images/aau-city-2.jpg')

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

ax[0].imshow(image1)
ax[1].imshow(image2)
```

```python
image1_gray = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
image2_gray = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

ax[0].imshow(image1_gray, cmap='gray')
ax[1].imshow(image2_gray, cmap='gray')
```

```python
sift = cv2.SIFT_create()

keypoints1, descriptors1 = sift.detectAndCompute(image1_gray, None)
keypoints2, descriptors2 = sift.detectAndCompute(image2_gray, None)

fig, ax = plt.subplots(2, 1, figsize=(20, 5))

ax[0].imshow(descriptors1.T)
ax[1].imshow(descriptors2.T)
```

```python
keypoint_image1 = cv2.drawKeypoints(image1, keypoints1, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
keypoint_image2 = cv2.drawKeypoints(image2, keypoints2, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

fig, ax = plt.subplots(1, 2, figsize=(20, 10))

ax[0].imshow(keypoint_image1)
ax[1].imshow(keypoint_image2)
```
