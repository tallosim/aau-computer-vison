```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from helpers import *
```

```python
img = cv2.imread('Images/paper.jpg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(img.shape)
show_histogram(img)
plt.imshow(img, cmap='gray')
```

```python
A = 1
B = 10
T = 30
BLUR_SIZE = 130

new_img = cv2.blur(img, (BLUR_SIZE,BLUR_SIZE))
new_img = contrast_brightness(new_img, A, B)
new_img = np.abs(new_img - img)
show_histogram(new_img)
new_img = cv2.threshold(new_img, T, 255, cv2.THRESH_BINARY_INV)[1]

print(new_img.shape)
plt.imshow(new_img, cmap='gray')
```
