```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from helpers import *
```

```python
img = cv2.imread('Images/low-contrast.jpeg', 1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

h, w = img.shape
h, w
```

```python
plt.imshow(img, cmap='gray')
```

```python
show_histogram(img)
```

```python
a = 1.8
b = -90
new_image = np.zeros((h, w), dtype=np.uint8)

for i in range(h):
    for j in range(w):
        new_image[i, j] = minmax(a * img[i, j] + b)
        
plt.imshow(new_image, cmap='gray')
```

```python
show_histogram(new_image)
```
