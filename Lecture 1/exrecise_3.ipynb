```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from helpers import *
```

```python
color_img = cv2.imread('Images/color.jpeg', 1)
color_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)

color_img = np.array(color_img)

plt.imshow(color_img)
color_img.shape
```

```python
show_histograms(color_img)
```

```python
#1. Using nested loops
gray_scale = np.zeros(color_img.shape, dtype=np.uint8)

for x in range(color_img.shape[0]):
    for y in range(color_img.shape[1]):
        gray_scale[x, y] = np.average(color_img[x, y])
        
plt.imshow(gray_scale, cmap='gray')
```

```python
#2. Using matrix multiplication
gray_scale = color_img @ [0.299, 0.587, 0.114]

plt.imshow(gray_scale, cmap='gray')
```

```python
#3. Using the built-in OpenCV function
gray_scale = cv2.cvtColor(color_img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_scale, cmap='gray')
```
