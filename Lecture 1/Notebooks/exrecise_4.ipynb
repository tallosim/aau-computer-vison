```python
T = 120

mask_1 = cv2.threshold(thermal_1_img, T, 255, cv2.THRESH_BINARY)[1]
mask_2 = cv2.threshold(thermal_2_img, T, 255, cv2.THRESH_BINARY)[1]

and_mask = cv2.bitwise_and(mask_1, mask_2)
or_mask = cv2.bitwise_or(mask_1, mask_2)
```
