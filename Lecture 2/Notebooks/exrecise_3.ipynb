```python
import cv2
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import numpy as np
from helpers import *
```

```python
image = cv2.imread('../Images/dots.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

show_histograms(image)
plt.imshow(image, cmap='gray')
```

```python
T = 80

mask = cv2.threshold(image, T, 255, cv2.THRESH_BINARY_INV)[1]

plt.imshow(mask, cmap='gray')
```

```python
new_mask = cv2.medianBlur(mask, 3)

plt.imshow(new_mask, cmap='gray')
```

```python
new_mask = np.copy(mask)

# new_mask = cv2.medianBlur(new_mask, 3)
new_mask = erosion(new_mask, 1)
new_mask = cv2.medianBlur(new_mask, 3)
new_mask = dilation(new_mask, 2)

plt.imshow(new_mask, cmap='gray')
```
