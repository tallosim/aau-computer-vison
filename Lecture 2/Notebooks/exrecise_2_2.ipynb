```python
import cv2
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import numpy as np
from helpers import *
```

```python
image = cv2.imread('../Images/neon-text.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

show_histograms(image)
plt.imshow(image)
```

```python
x = 276
y = 193
w = 200
h = 190

fig, ax = plt.subplots()

ax.imshow(image)
ax.add_patch(patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='b', facecolor='none'))

plt.show()
```

```python
template = image[y:y+h, x:x+w]

plt.imshow(template)
```

```python
template_matching_image = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)

plt.imshow(template_matching_image)
```

```python
T = 0.99

mask = template_matching_image > T

plt.imshow(mask, cmap='gray')
```
