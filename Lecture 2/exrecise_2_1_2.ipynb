```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from helpers import *
```

```python
image = cv2.imread('Images/lion.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

show_histograms(image)
plt.imshow(image)
```

```python
KERNEL_RADIUS = 3

kernel = genarate_gussian_kernel(KERNEL_RADIUS)
new_image = np.zeros(image.shape, dtype=np.uint8)

for x in range(KERNEL_RADIUS, image.shape[0] - KERNEL_RADIUS):
    for y in range(KERNEL_RADIUS, image.shape[1] - KERNEL_RADIUS):
        for channel in range(image.shape[2]):
            temp = 0
            for i in range(-KERNEL_RADIUS, KERNEL_RADIUS + 1):
                for j in range(-KERNEL_RADIUS, KERNEL_RADIUS + 1):
                    temp += image[x + i, y + j, channel] * kernel[i + KERNEL_RADIUS, j + KERNEL_RADIUS]
            new_image[x, y, channel] = int(temp)

show_histograms(new_image)
plt.imshow(new_image)
```

```python
KERNEL_RADIUS = 3

kernel = genarate_gussian_kernel(KERNEL_RADIUS)
new_image = cv2.filter2D(image, -1, kernel)

show_histograms(new_image)
plt.imshow(new_image)
```

```python
KERNEL_RADIUS = 3

new_image = cv2.GaussianBlur(image, (KERNEL_RADIUS * 2 + 1, KERNEL_RADIUS * 2 + 1), 1, 1)

show_histograms(new_image)
plt.imshow(new_image)
```
